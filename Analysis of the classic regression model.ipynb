```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
from scipy.stats import t
import statsmodels.api as sm

n = 1000
x_i = np.linspace(0, 10, n)
b_0 = 3
b_1 = 2
sigma = 1
N = 1000

def Monte_carlo(N, sigma, n, b_0, b_1):
    beta_0_st = []
    beta_1_st = []

    for i in range(N):
        x_i = np.linspace(0, 10, n)
        eps_i = np.random.normal(0, 1, n)
        Y_i = b_0 + b_1 * x_i + eps_i
        x_sr = np.mean(x_i)
        Y_sr = np.mean(Y_i)

        beta_1 = np.sum((x_i - x_sr) * Y_i) / np.sum((x_i - x_sr) ** 2)
        beta_0 = Y_sr - beta_1 * x_sr
        Y_est = beta_0 + beta_1 * x_i
        s = np.sqrt(np.sum((Y_i - Y_est) ** 2) / (n - 2))

        beta_0_st.append((beta_0 - b_0) / (s * np.sqrt((1 / n) + ((x_sr ** 2) / np.sum((x_i - x_sr) ** 2)))))
        beta_1_st.append((beta_1 - b_1) / (s / np.sqrt(np.sum((x_i - x_sr) ** 2))))

    return beta_0_st, beta_1_st

Dane = Monte_carlo(N, sigma, n, b_0, b_1)

b_0_dist = Dane[0]
x1 = np.linspace(np.min(b_0_dist), np.max(b_0_dist), 1000)
#x1 = np.linspace(-10,10,1000)
ecdf = sm.distributions.ECDF(b_0_dist)

y = ecdf(x1)
y_theo = st.t.cdf(x1, df=n-2)

plt.plot(x1, y, label='emp')
plt.plot(x1, y_theo, label='theo')
plt.title("Porwnanie dystrybuant empirycznej i teoretycznej dla b_0")
plt.xlabel("x")
plt.ylabel("warto")
plt.legend()
plt.show()

b_1_dist = Dane[1]
x2 = np.linspace(np.min(b_1_dist), np.max(b_1_dist), 1000)
ecdf1 = sm.distributions.ECDF(b_1_dist)

y1 = ecdf1(x2)
y1_theo = st.t.cdf(x2, df=n-2)

plt.plot(x2, y1, label='emp')
plt.plot(x2, y1_theo, label='theo')
plt.title("Porwnanie dystrybuant empirycznej i teoretycznej dla b_1")
plt.xlabel("x")
plt.ylabel("warto")
plt.legend()
plt.show()
```

```python
def rezim_r(N, n, r, sig_1, sig_2):
    # r to teoretyczny punkt reimu
    X = np.linspace(0,n,n)
    BK = np.zeros(N)
    for i in range(N):
        print(i)
        Y = np.zeros(n)
        # np.concatenate
        for a in range(r):
            Y[a] = st.norm.rvs(loc=0, scale=sig_1, size=1)
        for a in range(r,n):
            Y[a] = st.norm.rvs(loc=0, scale=sig_2, size=1)
        # cumsum
        CY = np.cumsum(Y**2)
        plt.plot(X, CY)
        
        bk = np.zeros(n-3)
        for k in range(n-3):
            # dopasowanie wspczynnikw prostej 1 i 2
            
            a_1, b_1 = np.polyfit(X[2:k+3], CY[2:k+3], 1)
            a_2, b_2 = np.polyfit(X[k+3:], CY[k+3:], 1)
            
            bk[k] = np.sum((CY[2:k+3]-a_1*X[2:k+3]-b_1)**2) + np.sum((CY[k+3:]-(a_2*X[k+3:]+b_2))**2)
        for a in range(len(bk)):
            if bk[a] == min(bk):
                BK[i] = a

    return BK


n = 2000
r = 800
```

```python
data_1 = rezim_r(100,2000,800,2,3)
data_2 = rezim_r(100,2000,800,2,4)
data_3 = rezim_r(100,2000,800,2,10)
```
